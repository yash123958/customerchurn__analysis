```python
def con(value):
    if value==1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(con)      
    
```

```python
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt

ax=sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title("Count Number of Churned Customers")
plt.show()
```

Most customers did not churn (5174) compared to 1869 customers who churned, showing customer retention is higher.
The churned customers are significantly fewer, indicating around 25–30% of customers left the service.
Since churn still exists in a noticeable amount, the company should analyze reasons behind the 1869 churned customers to improve retention.

```python
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%", colors=['green','yellow'])
plt.title('Percentage Of Churned Customers', fontsize=10)
plt.show()
```

```python
ax=sns.countplot(x="gender",data=df,hue="Churn",palette=['purple','red'])
ax.bar_label(ax.containers[0])
plt.title("Count of Churned Customers according to Gender")
plt.show()
```

Male customers who did not churn (2625) are slightly higher than female customers who did not churn (2549).
The number of customers who churned is almost the same for both genders (around 930–940).
This suggests that gender has very little impact on customer churn, and other factors may influence churn more.

```python
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="SeniorCitizen",data=df,hue="Churn",palette=['yellow','green'])
plt.show()
```

Most customers are not senior citizens, and the majority of them did not churn, indicating better retention in this group.
Senior citizens have a relatively higher churn proportion compared to non-senior customers, meaning they leave the service more frequently.
Although the number of senior citizens is small, a noticeable portion of them churned, suggesting this group may require targeted retention strategies.
Overall, non-senior customers dominate the dataset, but churn risk appears higher among senior citizens.

```python
ax=pd.crosstab(df["SeniorCitizen"],df["Churn"]).plot(kind="bar",stacked=True)
ax.bar_label(ax.containers[0])
plt.show()
```

Non-senior citizens form the majority of customers (around 4508 did not churn), showing that most users in the dataset are not senior citizens.
Churn is present in both groups, but the number of churned customers is higher among non-senior citizens because this group has a much larger total population.
Senior citizens have fewer total customers, but a significant portion of them churn, indicating relatively higher churn risk compared to their group size.
The chart suggests that senior citizen customers may require special retention strategies, even though they represent a smaller segment of the customer base.

```python
sns.histplot(x="tenure",data=df,hue="Churn",palette=['yellow','green'])
plt.title("Distribution of Tenure according to Churn")
plt.show()
```

Customers with low tenure (0–10 months) have the highest churn, indicating that new customers are more likely to leave the service early.
As tenure increases, the number of churned customers decreases, meaning long-term customers are more likely to stay.
Customers with high tenure (50–72 months) mostly did not churn, showing strong loyalty among long-term users.
Overall, tenure has a strong relationship with churn — customers who stay longer with the company are less likely to leave.

```python
ax=sns.countplot(x="Contract",data=df,hue="Churn",palette=['cyan','magenta'])
ax.bar_label(ax.containers[0])
plt.title("Count of Churned Customers according to Contract Type")
plt.show()
```

Month-to-month contracts have the highest churn, with a large number of customers leaving compared to other contract types.
Customers with one-year contracts churn much less, showing that longer commitments improve customer retention.
Two-year contracts have the lowest churn rate, meaning customers with long-term contracts are most likely to stay.
Overall, longer contract duration reduces churn, indicating that contract commitment plays an important role in customer retention.

```python
sns.lineplot(x="tenure", y="MonthlyCharges", data=df, errorbar=None,marker="o")

plt.grid(True)
plt.show()
```

Monthly charges generally increase with tenure, showing a positive relationship between how long a customer stays and how much they pay.
New customers (low tenure) tend to have lower monthly charges, possibly due to basic plans or promotional offers.
Long-term customers (higher tenure) show higher monthly charges, which may indicate upgraded plans or additional services over time.
Although the trend is increasing, there are small fluctuations, meaning monthly charges vary slightly across different tenure periods.

```python
sns.kdeplot(x="MonthlyCharges", hue="Churn", data=df, fill=True,palette=['cyan','magenta'])

plt.title("Monthly Charges Distribution by Churn")
plt.show()
```

Customers with higher monthly charges are more likely to churn, as the churn (Yes) density is higher in the 70–100 monthly charges range.
Customers with lower monthly charges (around 20–40) mostly did not churn, indicating that cheaper plans have better customer retention.
The non-churn group has a wider distribution, meaning customers who stay are spread across many pricing ranges.
Overall, higher monthly charges appear to increase the probability of churn, suggesting price sensitivity among customers.

```python
sns.kdeplot(x="TotalCharges", hue="Churn", data=df, fill=True,palette=['red','purple'])

plt.title("Monthly Charges Distribution by Churn")
plt.show()
```

Customers with low total charges show higher churn, indicating that newer or short-term customers are more likely to leave.


Customers with high total charges mostly did not churn, suggesting that long-term customers are more loyal to the service.


The non-churn group has a wider spread across higher total charges, meaning customers who stay tend to accumulate larger total payments over time.


Overall, higher total charges are associated with longer customer tenure and lower churn probability.
